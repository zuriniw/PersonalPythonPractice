## ⭐Dictionary basic knowledge

>字典（Dictionary）是一种用于存储**键-值对（Key-Value）数据**的数据结构。

它也被称为关联数组、哈希表或映射.
字典中的每个元素由一个key和一个对应的value组成，
key是唯一的，且不可变的，而value可以是任意类型的数据。

字典的特点是快速查找（用hash table技术），所以检索只需要用key对应到value，而不需要遍历
key是唯一的，不可变的（这也意味着key不能是可变的list，但可以是不变的tuple）



```python

my_dict = {"name": "John", "age": 25, "city": "New York"}
```

在上面这个例子中`my_dict` 是一个字典，它包含了三个键值对Key-Value:
- key `"name"` 对应的value是 `"John"`
- key `"age"` 对应的value是 `25`
- key `"city"` 对应的value是 `"New York"`

```python
def count_letters(word_string):
    """ 
    This function takes as input a string composed entirely of lower case letters.
    It returns the lower case letter that appears most frequently in the string.
    In case of ties, it returns the earliest letter in alphabetical order.
    """
    
    # Initialize a dictionary with each letter of the alphabet as keys and 0 as their values
    ALPHABET = "abcdefghijklmnopqrstuvwxyz"
    letter_count = {letter: 0 for letter in ALPHABET}
    
    # Count the occurrence of each letter in the word list
    for char in word_string:
        if char in letter_count:  # Ensure the letter is in the alphabet
            letter_count[char] += 1
    
    # Find the letter with the maximum frequency
    max_count = max(letter_count.values())
    most_frequent_letters = [letter for letter, count in letter_count.items() if count == max_count]
    
    # Return the earliest letter in alphabetical order in case of ties
    return min(most_frequent_letters)

# The input should be a string of concatenated words, not a list of characters
monty_quote = "listen strange women lying in ponds distributing swords is no basis for a system of government supreme executive power derives from a mandate from the masses not from some farcical aquatic ceremony"
monty_words = monty_quote.replace(" ", "")  # remove spaces to get one continuous string of letters
    
# Call the function and print the result
print(count_letters(monty_words))  # Should return the most frequent letter
```

💡:

letter_count = {letter: 0 for letter in ALPHABET}

>ALPHABET 是一个字符串，包含了英文字母表中的所有小写字母。

>for letter in ALPHABET 是一个循环，它遍历 ALPHABET 中的每个字母。

>{letter: 0 for letter in ALPHABET} 创建了一个字典，其中每个字母 letter 都是一个键（key），并且每个键的初始值（value）都被设置为 0。

最终的结果是一个字典 letter_count，它的键是字母表中的每个字母，每个键对应的值都是 0。这个字典可以被用来计数每个字母在文本中出现的次数。

💡：

most_frequent_letters = [letter for letter, count in letter_count.items() if count == max_count]

>letter_count.items() 返回一个包含字典中所有键值对（key-value pairs）的迭代器。

>for letter, count in letter_count.items() 是一个循环，遍历上面提到的所有键值对，其中 letter 是字典中的键（key），count 是与之对应的值（value）。

>if count == max_count 是一个条件语句，它检查当前键的值（即字母出现的次数）是否等于之前找到的最大值 max_count。
