````python
"""
Iterating over lists.
"""

def print_items(alist):
    """
    Iterate over alist and print all items.
    """
    for item in alist:
        print(item)

def print_items_bad(alist):
    """
    Iterate over alist and print all items.
    List indexing in this way is error prone and unnecessary.
    """
    length = len(alist)
    for index in range(length):
        print(alist[index])

def count_items(alist):
    """
    Count number of items in alist.
    """
    count = 0
    for item in alist:
        count = count + 1
    return count

def count_odd_items(numlist):
    """
    Count number of odd numbers in numlist.
    """
    count = 0
    for num in numlist:
        if num % 2 == 1:
            count += 1
    return count

numbers = list(range(5, 82, 3))
strings = ["python", "is", "fun", "!"]

print("Iterate over lists and print items")
print("")

print_items(numbers)
print_items(strings)
print_items_bad(numbers)

print("")
print("Iterate over lists and process them")
print("")

print(count_items(numbers))
print(count_items(strings))
print(count_odd_items(numbers))
````

假设你有两个列表，一个是数字列表 `numbers`，另一个是宠物名字列表 `pets`。列表定义如下：

```python
numbers = [5, 2, 9, 1, 5, 6]
pets = ["dog", "cat", "rabbit"]
```

请完成以下任务：

1. 计算 `numbers` 列表中所有数字的平方和。
2. 打印出 `pets` 列表中每个宠物名称的长度。
3. 找出 `numbers` 列表中的最大数字，打印出 `"The largest number is [最大数]"`。
4. 为 `pets` 列表中的每个宠物名称添加 `" pet"` 后缀，然后打印更新后的列表。



````python
numbers = [5, 2, 9, 1, 5, 6]
pets = ["dog", "cat", "rabbit"]

##1
j = 0
for i in numbers:
    j = j + i ** 2
print(j)

##2
name_len = list((len(name) for name in pets))
print (name_len)

##3
max_number = max(numbers)
print (max_number)

##4
print(list((pet_name + " pet" for pet_name in pets)))
````
